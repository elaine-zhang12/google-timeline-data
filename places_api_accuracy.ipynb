```python
# Import libraries
import pandas as pd
from fractions import Fraction
```

```python
# Compare latitude and longitude tuples between places_info and cluster means - CHERYL

folder = "CB_data" # ONLY need to change this value

cluster_data_file_path = folder + "/cluster_means.csv"
places_data_file_path = folder + "/places_info.csv"

df_clusters = pd.read_csv(cluster_data_file_path)
df_places = pd.read_csv(places_data_file_path)

cluster_coords = []
places_coords = []

for index, row in df_clusters.iterrows():
    cluster_coords.append((row.latitude, row.longitude))

for index, row in df_places.iterrows():
    places_coords.append((row.latitude, row.longitude))

matching_coords = 0
for coord in cluster_coords:
    if coord in places_coords:
        matching_coords += 1

print("Number of matching coords: ", matching_coords)
print("Total cluster coords: ", len(cluster_coords))

print("The Places API accuracy is ", matching_coords/len(cluster_coords))
```
